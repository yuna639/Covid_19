```python
# ===================================================
# üìà Linear Regression Model on Amazon Product Data
# ===================================================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ------------------------------
# 1Ô∏è‚É£ Load and prepare dataset
# ------------------------------
df = pd.read_csv(r'process\cleaned_amazon_reviews.csv')  # Change filename if needed

# Some columns may be read as object because of stray characters (e.g. '|' in some cells).
# Coerce rating and rating_count to numeric, invalid parsing will become NaN.
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Drop rows that have NaNs in any of the features or targets we will use
required_cols = ['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']
df = df.dropna(subset=required_cols).reset_index(drop=True)

# Features and target variables
X = df[['actual_price', 'discounted_price', 'discount_percentage']]

# Two targets to analyze business relations
y_rating = df['rating'].astype(float)          # customer satisfaction
y_popularity = df['rating_count'].astype(float)  # product popularity

# ------------------------------
# 2Ô∏è‚É£ Train-test split
# ------------------------------
X_train, X_test, y_train_r, y_test_r = train_test_split(X, y_rating, test_size=0.3, random_state=42)
_, _, y_train_p, y_test_p = train_test_split(X, y_popularity, test_size=0.3, random_state=42)

# ------------------------------
# 3Ô∏è‚É£ Build Linear Regression Models
# ------------------------------
model_rating = LinearRegression()
model_popularity = LinearRegression()

model_rating.fit(X_train, y_train_r)
model_popularity.fit(X_train, y_train_p)

# ------------------------------
# 4Ô∏è‚É£ Evaluate performance
# ------------------------------
y_pred_r = model_rating.predict(X_test)
y_pred_p = model_popularity.predict(X_test)

def regression_summary(y_true, y_pred, name):
    print(f"\nüìò {name} MODEL PERFORMANCE")
    print("-" * 40)
    print(f"MAE: {mean_absolute_error(y_true, y_pred):.4f}")
    print(f"MSE: {mean_squared_error(y_true, y_pred):.4f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_true, y_pred)):.4f}")
    print(f"R¬≤ Score: {r2_score(y_true, y_pred):.4f}")

regression_summary(y_test_r, y_pred_r, "Customer Rating (Satisfaction)")
regression_summary(y_test_p, y_pred_p, "Product Popularity (Sales Proxy)")

# ------------------------------
# 5Ô∏è‚É£ Coefficient Analysis (Business Insights)
# ------------------------------
coef_df_rating = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model_rating.coef_,
    'Impact_on': 'Customer Rating'
})

coef_df_popularity = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model_popularity.coef_,
    'Impact_on': 'Product Popularity'
})

coef_df = pd.concat([coef_df_rating, coef_df_popularity])
print("\nüìä Coefficient Summary:\n", coef_df)

# ------------------------------
# 6Ô∏è‚É£ Visualization: Regression Lines
# ------------------------------

# A. Relationship: Discount % vs Rating
plt.figure(figsize=(8,5))
sns.regplot(x=df['discount_percentage'], y=df['rating'], scatter_kws={'s':60}, color='seagreen')
plt.title('H1: Discount Percentage vs Customer Rating')
plt.xlabel('Discount (%)')
plt.ylabel('Rating')
plt.show()

# B. Relationship: Discount % vs Popularity (Rating Count)
plt.figure(figsize=(8,5))
sns.regplot(x=df['discount_percentage'], y=df['rating_count'], scatter_kws={'s':60}, color='royalblue')
plt.title('H2: Discount Percentage vs Product Popularity')
plt.xlabel('Discount (%)')
plt.ylabel('Number of Ratings (Popularity)')
plt.show()

# C. Actual vs Predicted for Ratings
plt.figure(figsize=(8,5))
sns.scatterplot(x=y_test_r, y=y_pred_r, color='darkorange')
plt.plot([min(y_test_r), max(y_test_r)], [min(y_test_r), max(y_test_r)], 'r--')
plt.title('Predicted vs Actual: Customer Ratings')
plt.xlabel('Actual Ratings')
plt.ylabel('Predicted Ratings')
plt.show()

# D. Actual vs Predicted for Popularity
plt.figure(figsize=(8,5))
sns.scatterplot(x=y_test_p, y=y_pred_p, color='dodgerblue')
plt.plot([min(y_test_p), max(y_test_p)], [min(y_test_p), max(y_test_p)], 'r--')
plt.title('Predicted vs Actual: Product Popularity (Rating Count)')
plt.xlabel('Actual Popularity')
plt.ylabel('Predicted Popularity')
plt.show()

# ------------------------------
# 7Ô∏è‚É£ Pairplot for Deep Insights
# ------------------------------
sns.pairplot(df[['actual_price','discounted_price','discount_percentage','rating','rating_count']],
             diag_kind='kde', corner=True, plot_kws={'alpha':0.7})
plt.suptitle('üìà Pairwise Relationships Among Key Business Variables', y=1.02, fontsize=18)
plt.show()

# ------------------------------
# 8Ô∏è‚É£ Business Interpretation Summary
# ------------------------------
print("\nüí° BUSINESS INSIGHTS:")
print("1Ô∏è‚É£ Discount percentage has a positive effect on both ratings and popularity ‚Üí discounts attract engagement.")
print("2Ô∏è‚É£ Actual price has a weak or negative correlation with ratings ‚Üí customers expect higher quality for higher prices.")
print("3Ô∏è‚É£ Discounted price remains a key driver of sales conversion ‚Äî affordable products gain more attention.")
print("4Ô∏è‚É£ Model helps identify ideal pricing-discount combinations to maximize both sales and satisfaction.")
```
